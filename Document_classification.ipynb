# Document Classification using CNN
1. Problem Statement

Financial institutions require automated systems to verify and classify customer documents such as Aadhaar, PAN, and cheque images. Manual verification is time-consuming and error-prone. This module aims to develop a Convolutional Neural Network (CNN) model to classify uploaded document images and detect invalid inputs.

2. Dataset Description

A custom dataset was created consisting of images of:
- Aadhaar cards
- PAN cards
- Cheques
- Other unrelated images

Each class contains approximately 15–20 images. Data augmentation techniques were used to increase dataset diversity and improve model performance.

3. Objectives

- To build a CNN model for document classification
- To classify documents into Aadhaar, PAN, cheque, and others
- To detect invalid or unrelated images
- To simulate document verification in financial systems

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

#step2 : Load Dataset 

```python
img_size = (128, 128)
batch_size = 32

train_data = tf.keras.preprocessing.image_dataset_from_directory(
    "dataset",
    image_size=img_size,
    batch_size=batch_size,
    validation_split=0.2,
    subset="training",
    seed=42
)

val_data = tf.keras.preprocessing.image_dataset_from_directory(
    "dataset",
    image_size=img_size,
    batch_size=batch_size,
    validation_split=0.2,
    subset="validation",
    seed=42
)
```

#step3 : Get Class Names

```python
class_names = train_data.class_names
print(class_names)
```

#step4 : Normalize Data

```python
normalization_layer = layers.Rescaling(1./255)

train_data = train_data.map(lambda x, y: (normalization_layer(x), y))
val_data = val_data.map(lambda x, y: (normalization_layer(x), y))
```

#step5 : Build CNN model

```python
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    layers.MaxPooling2D(),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(class_names), activation='softmax')
])
```

#step6 : Compile Model

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

#step7 : Train Model

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

#step8 : Evaluate 

```python
loss, accuracy = model.evaluate(val_data)
print("Validation Accuracy:", accuracy)
```

#step9 : Save Model

```python
model.save("document_classifier.h5")
```

#step10 : Testing 

```python
import numpy as np
from tensorflow.keras.preprocessing import image

img = image.load_img("test_images/test1.jpg", target_size=(128,128))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0

prediction = model.predict(img_array)

confidence = np.max(prediction)
predicted_class = class_names[np.argmax(prediction)]

if confidence < 0.7:
    print("❌ Invalid or unclear document")
else:
    print("Prediction:", predicted_class)
    print("Confidence:", confidence)
```

5. Methodology

- Images are resized to 128x128 pixels
- Data is normalized for efficient training
- A Convolutional Neural Network (CNN) is used for feature extraction
- The model is trained using training and validation splits
- A confidence threshold is used to detect invalid inputs  

6. Result

The model achieved a validation accuracy of approximately 70%. It successfully classifies document types and detects invalid inputs using confidence thresholding.
